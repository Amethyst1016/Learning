### 53 CountVectorizer

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

documents = ['python is a programming language',
              'python is popular',
              'programming in python',
              'object-oriented prigramming in python']
```

```python
vectorizer = CountVectorizer()
vectorizer.fit_transform(documents)
df = pd.DataFrame(vectorizer.fit_transform(documents).toarray(),
                  columns = vectorizer.get_feature_names_out())
df
```

### 54 set stopword

```python
vectorizer = CountVectorizer(stop_words='english')
vectorizer.fit_transform(documents)
df = pd.DataFrame(vectorizer.fit_transform(documents).toarray(),
                  columns = vectorizer.get_feature_names_out())
df
```

### 55 set n-gram

```python
vectorizer = CountVectorizer(stop_words='english', ngram_range=(1,2))
vectorizer.fit_transform(documents)
df = pd.DataFrame(vectorizer.fit_transform(documents).toarray(),
                  columns = vectorizer.get_feature_names_out())
df
```

### 56 TF-IDF (term frequency-inverse document frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorize = TfidfVectorizer()
df = pd.DataFrame(tfidf_vectorize.fit_transform(documents).toarray(),
                  columns = tfidf_vectorize.get_feature_names_out())
df
```

### 57 set stopword

```python
tfidf_vectorize = TfidfVectorizer(stop_words='english') # stop_words=['is','in']
df = pd.DataFrame(tfidf_vectorize.fit_transform(documents).toarray(),
                  columns = tfidf_vectorize.get_feature_names_out())
df
```
