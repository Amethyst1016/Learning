### 08 discretization - equal width interval

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'weight': [75,78.5,85,91,84.5,83,68]})
df
```

```python
df['weight_cut'] = pd.cut(df['weight'], bins=3) # 3 intervals, type=category
df
```

### 09 discretization - custom interval

```python
df['weight_cut_custom'] = pd.cut(df['weight'], bins=[60,75,80,95])
# adjacent element form a interval, type=category
df
```

### 10 discretization - tag each interval

```python
df['weight_cut_label'] = pd.cut(df['weight'], bins=[60,75,80,95], labels=['light','normal','heavy'])
df
```

### 11 discretization - dummies

```python
df = df.drop(['weight_cut','weight_cut_custom'], axis=1)
df = pd.get_dummies(df)
df
```
