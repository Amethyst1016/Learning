```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# load data
oecd_bli = pd.read_csv('datasets/lifesat/oecd_bli_2015.csv', thousands=',')
gdp_per_capita = pd.read_csv('datasets/lifesat/gdp_per_capita.csv', thousands=',', delimiter='\t',
                            encoding='latin1', na_values='n/a')
display(oecd_bli.head())
display(gdp_per_capita.head())
```

```python
# preprocess data
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"] # filter
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value") # reshape df
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp_per_capita.set_index("Country", inplace=True)

display(oecd_bli.head())
display(gdp_per_capita.head())
```

```python
full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                              left_index=True, right_index=True) # since index of two df is the same
display(full_country_stats.head())
```

```python
full_country_stats.sort_values(by="GDP per capita", inplace=True) # asc
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices)) # len(full_country_stats)=36, 36 countries in total
country_stats = full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices] # keep two col and 29 rows
display(country_stats.head()) 
```

```python
# split x and y
x = np.c_[country_stats['GDP per capita']]
y = np.c_[country_stats['Life satisfaction']]
# visualize data
plt.scatter(x,y)
plt.xlabel('GDP per capita')
plt.ylabel('Life satisfaction')
plt.show()
```

### model-based prediction (linear regression)

```python
# set linear model
model = LinearRegression()
# train model
model.fit(x,y)
# predict on new data
x_new=[[22587]]
model.predict(x_new)
```

### instance-based prediction (KNN)

```python
from sklearn.neighbors import KNeighborsRegressor
# set KNN model
model = KNeighborsRegressor(n_neighbors=3)
# train model
model.fit(x,y)
# predict on new data
x_new=[[22587]]
model.predict(x_new)
```
